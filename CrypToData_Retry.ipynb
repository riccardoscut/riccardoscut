```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt   # data visualization
pd.options.mode.chained_assignment = None  # suppress DataFrame slice warning
```

```python
df = pd.read_csv('dataset.csv')
```

```python
btc = df.loc[df['crypto_name'] == 'Bitcoin'].copy()
btc = btc.set_index('date')
btc.index = pd.to_datetime(btc.index)
btc
```

```python
# Calcolo dei minimi assoluti tra gli halving
halving_dates = [
    pd.Timestamp('2012-11-28'),
    pd.Timestamp('2016-07-09'),
    pd.Timestamp('2020-05-11'),
    pd.Timestamp('2024-04-19')  # Stima del prossimo halving
]

# Identificare i minimi assoluti tra gli halving
minima = []
for i in range(len(halving_dates)):
    start = btc.index.min() if i == 0 else halving_dates[i - 1]
    end = halving_dates[i] if i < len(halving_dates) else btc.index.max()
    
    # Controllo: Intervallo con dati
    data_in_range = btc.loc[start:end]
    if data_in_range.empty:
        print(f"Nessun dato tra {start} e {end}. Skipping.")
        continue
    
    # Calcolo del minimo
    minimum = data_in_range['close'].idxmin()
    minima.append(minimum)

# Creare i periodi dai minimi
periods = []
for i in range(len(minima) - 1):
    periods.append({'start': minima[i], 'end': minima[i + 1], 'halving': halving_dates[i]})

# Visualizzare i periodi
print(periods)

# Ordinare il dataset per data
btc = btc.sort_index()
btc

for i in range(len(halving_dates)):
    start = btc.index.min() if i == 0 else halving_dates[i - 1]
    end = halving_dates[i] if i < len(halving_dates) else btc.index.max()
    
    print(f"Intervallo {i}: {start} - {end}")
    print(btc.loc[start:end].head())  # Mostra i primi dati nell'intervallo
```

```python
# Aggiungere una colonna per il periodo
btc['period'] = None
for i, period in enumerate(periods):
    btc.loc[period['start']:period['end'], 'period'] = f"Period_{i+1}"

# Calcolare statistiche per ciascun periodo
statistics = []
for period_name, group in btc.groupby('period'):
    stats = {
        'period': period_name,
        'volatility_avg': group['daily_return'].std(),  # VolatilitÃ  media
        'volume_avg': group['volume'].mean(),          # Media dei volumi
        'marketcap_avg': group['marketCap'].mean(),    # Media del marketcap
    }
    statistics.append(stats)

# Visualizzare le statistiche per ogni periodo
print(statistics)
```
